# 03 | Particle Flow and High-Dimensional Scaling\nThis notebook compares Bootstrap PF, LEDH, and Kernel-PFF on runtime and consistency as dimension grows.

```python
import numpy as np\nimport matplotlib.pyplot as plt\n\ndims = np.array([16, 32, 64, 96, 128, 160, 192, 224])\nrmse_pf = np.array([1.3, 1.6, 1.9, 2.2, 2.45, 2.7, 2.95, 3.2])\nrmse_ledh = np.array([1.0, 1.2, 1.45, 1.7, 1.95, 2.2, 2.45, 2.7])\nrmse_kpf = np.array([0.9, 1.05, 1.25, 1.45, 1.65, 1.85, 2.05, 2.25])\n\nplt.plot(dims, rmse_pf, marker='o', label='Bootstrap PF')\nplt.plot(dims, rmse_ledh, marker='o', label='LEDH')\nplt.plot(dims, rmse_kpf, marker='o', label='Kernel-PFF')\nplt.title('Scaling comparison (illustrative)')\nplt.xlabel('Dimension')\nplt.ylabel('RMSE')\nplt.legend()\nplt.show()
```
