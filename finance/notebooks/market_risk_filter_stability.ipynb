# 02 | Nonlinear EKF/UKF Stress Tests\nThis notebook sets up a range-bearing model and compares EKF/UKF error and NEES consistency.

```python
import numpy as np\nimport matplotlib.pyplot as plt\n\nsteps = np.arange(1, 61)\nekf_nees = 3.6 * np.exp(-steps / 30) + 1.2\nukf_nees = 2.6 * np.exp(-steps / 38) + 1.0\n\nplt.fill_between(steps, 0.9, 2.1, color='#d9f2e8', alpha=0.7, label='95% consistency band')\nplt.plot(steps, ekf_nees, label='EKF NEES')\nplt.plot(steps, ukf_nees, label='UKF NEES')\nplt.title('Illustrative NEES trend')\nplt.xlabel('Step')\nplt.ylabel('NEES')\nplt.legend()\nplt.show()
```
