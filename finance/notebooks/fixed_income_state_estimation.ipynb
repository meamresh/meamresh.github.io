# 01 | Linear-Gaussian Baseline
This notebook benchmarks standard Kalman filtering in the linear-Gaussian setting with Riccati vs. Joseph covariance updates.

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(7)
T = 100
x = np.zeros(T)
z = np.zeros(T)
for k in range(1, T):
    x[k] = x[k-1] + rng.normal(0, 0.4)
    z[k] = x[k] + rng.normal(0, 0.9)

plt.plot(x, label='true')
plt.plot(z, alpha=0.7, label='obs')
plt.title('Linear-Gaussian sequence')
plt.legend()
plt.show()
```
